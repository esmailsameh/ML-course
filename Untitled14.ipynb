```python
sns.kdeplot(data=data , x='Age', hue='Sleep Disorder')

plt.xlabel('Age')
plt.ylabel('Density')

plt.show()
```

```python
temp_data = []

for index, row in data.iterrows():
    gender = row['Gender']
    disorder = row['Sleep Disorder']
    
    if not any(x[0] == gender and x[1] == disorder for x in temp_data):
        temp_data.append([gender, disorder, 0])
    
    for item in temp_data:
        if item[0] == gender and item[1] == disorder:
            item[2] += 1

display(temp_data)
```

```python
data1 = pd.DataFrame(temp_data, columns=['Gender', 'Disorder', 'Count'])

sns.barplot(x='Gender', y='Count', hue='Disorder', data=data1)

plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

```python
temp_data = []

for index, row in data.iterrows():
    gender = row['BMI Category']
    disorder = row['Sleep Disorder']
    
    if not any(x[0] == gender and x[1] == disorder for x in temp_data):
        temp_data.append([gender, disorder, 0])
    
    for item in temp_data:
        if item[0] == gender and item[1] == disorder:
            item[2] += 1

display(temp_data)
```

```python
data2 = pd.DataFrame(temp_data, columns=['BMI Category', 'Sleep Disorder', 'Count'])

sns.barplot(x='BMI Category', y='Count', hue='Sleep Disorder', data=data2)

plt.xlabel('BMI Category')
plt.ylabel('Count')

plt.show()
```
